```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1) # Number of rows,number of columns, plot number you're referring to 
plt.plot(x,y,'r') #color red

plt.subplot(1, 2, 2)
plt.plot(y,x, 'b')
```

```python
# OOP
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width , height
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.5, 0.3]) #20 percent from left, 50 percent from the bottom
# 40% canvas width, 30% canvas height

axes1.plot(x,y)
axes2.plot(y,x)
```
