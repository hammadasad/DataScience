# Algorithm

The K-Means Algorithm
- Choose # of K (clusters)
- Randomly assign each vector/point to a cluster
- Until clusters stop changing, repeat the process:
    - For each cluster
        - Compute cluster centroid by taking the mean vector of points in the cluster
        - Assign each data point to the cluster for which the centroid is the closest
        
- Selecting the hyperparameter K:
    - One way is the elbow method
    - Compute Sum of Squared Error for some values of k (ie. 2, 4, 6, 8)
    - Sum of Squared Error is defined by the sum of the squared distance between each member of the cluster and its centroid
    - If you plot K against the SSE, you will see that the error decreases as K gets larger, this is because when the number of clusters increases, they should be smaller, so distortion is also smaller. 
    
- The idea of the elbow method is to choose the K at which the Sum of Squared Error decreases greatly

# Generate Artificial Data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

```python
# So it's a tuple, we're gonna plot these
# If you grab the 1st element, you get a NumPy array
data
```

```python
# Grab all the rows in the first column, and the rows in the second column
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
# We can see the cluster each data point belongs to
data[1]
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
# Remember we're going to start off by randomely assigning each point to a cluster
# and then find the centroid of each cluster
# It then reassigns the data points to the cluster who's centroid is the closest
# It calculates the new centroid for each cluster and repeats this over and over until cluster
# variation cannot be reduced any further
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
cluster_points = kmeans.cluster_centers_
```

```python
cluster_points
```

```python
kmeans.labels_
```

```python
# Let's compare our plots
# Create two plots
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('Original')
ax1.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
ax2.set_title('K-Means')
ax2.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap="rainbow")
```
